# Univariate Analysis



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution plot for the Price variable
sns.displot(df['Price'], kde=True, bins=30, color='blue') # Using displot
plt.title('Distribution of Laptop Prices')
plt.xlabel('Price')
plt.ylabel('No of Counts')
plt.show()
```

```python
# Dristribution Plot for screen size and Price

plt.figure(figsize=(15, 5))
sns.histplot(df['OpSys'], kde=True, bins=10, color='red')
plt.title('Distribution of Operating System')
plt.xlabel(' Operating System')
plt.ylabel('No of Counts')
plt.show()
```

**sns.histplot(df['OpSys'], kde=True, bins=10, color='red'): This is the core of the plot creation.**

* sns.histplot() is a function from the seaborn library used to create histograms.
* df['OpSys']: This specifies the data to be plotted  the OpSys column from the df dataframe.
* kde=True: This adds a Kernel Density Estimate (KDE) line to the histogram, providing a smoother representation of the data distribution.
* bins=10: This divides the data into 10 bins for the histogram. You can adjust this number to change the granularity of the visualization.

# Bivariate Analysis

```python
# create a bivariate analysis between price and company

plt.figure(figsize=(15, 5))
sns.boxplot(x='Company', y='Price', data=df,hue="OpSys")
plt.title('Price Distribution by Company')
plt.xlabel('Company')
plt.ylabel('Price')
plt.xticks(rotation=50)
plt.legend(title="OpSys", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

**plt.legend(title="OpSys", bbox_to_anchor=(1.05, 1), loc='upper left'): Adds a legend to the plot.**

* title="OpSys" sets the title of the legend to "OpSys."
* bbox_to_anchor=(1.05, 1) and loc='upper left' control the position of the legend, placing it outside the plot area to the top right.

```python
# Sort the DataFrame by 'Inches' in ascending order
df_sorted = df.sort_values(by=['Inches'], ascending=True)

# Creating a Bivariate analysis between Weight and Inches
plt.figure(figsize=(15, 5))
sns.scatterplot(y='Weight', x='Inches', data=df_sorted, hue="Company")  # Use the sorted DataFrame
plt.title('Laptop Weight vs. Screen Size')
plt.ylabel('Weight (kg)')
plt.xlabel('Screen Size (Inches)')
plt.xticks(rotation=50)
plt.legend(title="Company", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

#df_sorted = df.sort_values(by=['Inches'], ascending=True): This line is responsible for sorting the data.
* df: This is your original Pandas DataFrame containing the laptop data.
* sort_values(): This is a function that sorts the DataFrame.
* by=['Inches']: We are sorting the data based on the values in the Inches column (which likely represents the screen size of the laptops).
* ascending=True: This ensures that the data is sorted in ascending order (from smallest to largest screen size).
* df_sorted: The sorted DataFrame is stored in a new variable called df_sorted.

**plt.xticks(rotation=50): Rotates the labels on the x-axis by 50 degrees to make them more readable if they are long.**

# Multivariant Analysis

```python
# Replace '?' with NaN
df.replace('?', np.nan, inplace=True)

# Convert 'Price', 'Weight', and 'Inches' to numeric, handling errors
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')
df['Inches'] = pd.to_numeric(df['Inches'], errors='coerce')

# Now you can create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Price', 'Weight', 'Inches']].corr(), annot=True, cmap='plasma', fmt=".2f")
plt.title('Correlation Heatmap of Price, Weight, and Inches')
plt.show()
```

**The errors='coerce' argument ensures that any values that cannot be converted to numbers are replaced with NaN.**

* df[['Price', 'Weight', 'Inches']].corr(): This part calculates the correlation between the 'Price', 'Weight', and 'Inches' columns using the .corr() method. Correlation measures how strongly these variables are related to each other.
* annot=True: This displays the correlation values within the heatmap squares, making it easier to interpret.
* cmap='plasma': This sets the color map of the heatmap.

```python
# Create the pair plot, including 'Company' in the data subset
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df[['Price', 'Weight', 'Inches', 'Company']], diag_kind='kde', hue="Company") # Include 'Company' in the data
plt.suptitle('Pair Plot of Price, Weight, and Inches', y=1.02)
plt.legend(title="Company", bbox_to_anchor=(1.05, 1), loc='upper right')
plt.show()
```

**diag_kind='kde': This specifies that the diagonal plots (which usually show the distribution of a single variable) should be Kernel Density Estimation plots, providing a smoother representation of the data distribution.**

# Feature Engineering

**Analysis and Interpretation Steps**


```python
def extract_memory_features(Memory_info):
  if pd.isnull(Memory_info):
    return np.nan
  parts=Memory_info.split()

  if"SSD" in Memory_info: return " SSD"
  elif"HDD" in Memory_info: return " HDD"
  elif"Flash" in Memory_info: return " Flash"
  elif"Hybrid" in Memory_info: return " Hybrid"
  else: return "Other"

df["Memory_Type"]=df["Memory"].apply(extract_memory_features)
print(df["Memory_Type"].value_counts())
```

```python
# Average price by processor generation
avg_price_by_generation =df.groupby("Memory_Type")["Price"].mean()
print("Average Price by Processor Generation:")
print(avg_price_by_generation)
print("*"*50)
# Average price by brand
avg_price_by_brand = df.groupby("Company")["Price"].mean().sort_values(ascending=False)
print("\nAverage Price by Brand:")
print(avg_price_by_brand)
print("*"*50)
# Average price by brand
avg_price_by_TypeName = df.groupby("TypeName")["Price"].mean().sort_values(ascending=False)
print("\nAverage Price by TypeName:")
print(avg_price_by_TypeName)

```
