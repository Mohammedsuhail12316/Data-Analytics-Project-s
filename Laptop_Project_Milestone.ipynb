EDA Steps:

1.Univariate Analysis:

Analyze distributions and summary statistics for key variables like price, RAM, and screen size

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution plot for the Price variable
plt.figure(figsize=(8, 5))
sns.histplot(df['Price'], kde=True, bins=30, color='blue')
plt.title('Distribution of Laptop Prices')
plt.xlabel('Price')
plt.ylabel('No of Counts')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution plot for the Screen Size (Inches) variable
plt.figure(figsize=(30, 10))
sns.histplot(df['Inches'], kde=True, bins=20, color='Red')
plt.title('Distribution of Laptop Screen Sizes')
plt.xlabel('Screen Size (Inches)')
plt.ylabel('No of Counts')
plt.show()
```

Heres an inference based on the analysis goals:

1.Distribution of Laptop Screen Sizes:


*   Screen sizes are typically measured diagonally in inches, and the distribution could show patterns such as clustering around standard sizes like 13", 14", 15.6", and 17".
*   If the distribution is unimodal and centered around a specific size, it indicates that a particular screen size is most popular among laptops.

*   If the distribution is spread out, it suggests a variety of screen sizes are in demand, possibly catering to different consumer needs like portability, gaming, or professional work.

2.Distribution of Prices (Univariate Analysis):


*   Laptop prices usually exhibit a right-skewed distribution, with most laptops falling in an affordable price range and fewer laptops in the premium segment.
*   A multimodal distribution might indicate distinct market segments, such as budget, mid-range, and high-end laptops.

2.Bivariate Analysis:
Explore relationships between variables, e.g., price vs. RAM, weight vs. screen size.






```python

#1. Scatter plot of Price vs. RAM
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Ram", y="Price", hue="Company")
plt.title("Price vs. RAM")
plt.xlabel("RAM (GB)")
plt.ylabel("Price")
plt.legend(title="Company", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### **Inference for Scatter Plot of Price vs. RAM:**

1. Laptops with higher RAM generally have higher prices, showing a positive correlation.  
2. Budget laptops (4GB-8GB), mid-range (8GB-16GB), and high-end (16GB+) may form distinct clusters.  
3. Outliers might represent premium laptops with lower RAM but advanced features or brand value.  



```python
# 2. Scatter plot of Weight vs. Screen Size
plt.figure(figsize=(15,5))
sns.scatterplot(data=df, x="Inches", y="Weight", hue="TypeName")
plt.title("Weight vs. Screen Size")
plt.xlabel("Screen Size (Inches)")
plt.ylabel("Weight (kg)")
plt.legend(title="Type Name", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

**Inference for Scatter Plot of Weight vs. Screen Size:**

1. Larger screen sizes generally correlate with heavier laptops.  
2. Lightweight laptops with smaller screens indicate ultraportables, while heavier ones with larger screens are gaming or workstation models.  
3. Outliers may include lightweight large-screen ultrabooks or rugged small-screen laptops.  
4. Weight increases non-linearly with screen size due to material and design innovations.  
5. Clear clusters can reflect distinct categories like ultraportables, general-purpose, and gaming laptops.  

3.Multivariate Analysis:

Investigate interactions among multiple features, such as how CPU, RAM, and GPU jointly affect price.

```python
# Correlation Matrix (for numerical features)
numerical_features = df.select_dtypes(include=['number'])  # Select numerical columns
correlation_matrix = numerical_features.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# Define the relevant columns
columns_of_interest = ['Price', 'Inches']  # Replace with actual column names from your dataset

# Select and clean the data
correlation_data = df[columns_of_interest].copy()
correlation_data = correlation_data.replace('?', np.nan)  # Replace '?' with NaN
correlation_data = correlation_data.apply(pd.to_numeric, errors='coerce')  # Convert to numeric

# Calculate the correlation matrix after dropping NaN values
correlation_matrix = correlation_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
plt.title('Correlation Heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# Ensure that both Ram and Weight are strings before applying string operations
df["Ram"] = df["Ram"].astype(str).str.replace("GB", "", regex=False)
df["Weight"] = df["Weight"].astype(str).str.replace("kg", "", regex=False)

# Convert to numeric after cleaning
df["Ram"] = pd.to_numeric(df["Ram"], errors="coerce")
df["Weight"] = pd.to_numeric(df["Weight"], errors="coerce")
df["Inches"] = pd.to_numeric(df["Inches"], errors="coerce")

# Drop rows with missing values in critical columns
df_cleaned = df.dropna(subset=["Price", "Ram", "Weight", "TypeName"])
```

```python


# Create a boxplot to visualize Price distribution across laptop types
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_cleaned, x="TypeName", y="Price", palette="Set3")
plt.title("Boxplot of Price Distribution by Laptop Type", fontsize=16)
plt.xlabel("Laptop Type", fontsize=12)
plt.ylabel("Price (in currency)", fontsize=12)
plt.xticks(rotation=0)
plt.yticks(rotation=10)
plt.tight_layout()
plt.show()
```

```python


# Price Distribution by Company
plt.figure(figsize=(12, 6))
sns.boxplot(x='Company', y='Price', data=df)
plt.title('Price Distribution by Company')
plt.xlabel('Company')
plt.ylabel('Price')
plt.xticks(rotation=35)
plt.show()


```

### **Inference for Box Plot of Price Distribution by Laptop Type:**

1. **Variation in Price:** The box plot shows significant variation in laptop prices across different types, reflecting the diversity of features and target audiences for each category.  
2. **Median Price Differences:** The median price varies considerably between categories, with high-performance types (e.g., gaming or workstation laptops) generally having a higher median price.

**Feature Engineering :**

```python
# Extract processor generation from the "CPU" column
def extract_generation(cpu_info):
    if "i3" in cpu_info: return "i3"
    elif "i5" in cpu_info: return "i5"
    elif "i7" in cpu_info: return "i7"
    elif "i9" in cpu_info: return "i9"
    else: return "Other"
df["Processor_Generation"] = df["Cpu"].apply(extract_generation)
print(df["Processor_Generation"].value_counts())
```

1. The code extracts processor generations (i3, i5, i7, i9) from the "CPU" column in the dataset.  
2. The **i7 processors** are the most common, with **485 entries**.  
3. The **i5 processors** are the second most frequent, with **377 entries**.  
4. The **"Other" category** has **201 entries**, likely representing non-Intel CPUs or unrecognized formats.  
5. The **i3 processors** are the least frequent, with **123 entries**.
